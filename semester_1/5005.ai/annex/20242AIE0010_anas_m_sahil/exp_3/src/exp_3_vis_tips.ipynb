```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv("data/tips.csv")
scatter = plt.scatter(tips['day'], tips['tip'])
plt.colorbar(scatter, label='Total Bill')
```

```python
plt.plot(tips["day"])
plt.plot(tips["size"])
```

```python
plt.hist(tips["total_bill"])
```

```python
# Kernel Density Estimate
sns.histplot(x="total_bill", kde=True, hue="sex", data=tips)
```

```python
sns.lineplot(x="sex", y="total_bill", data=tips)
```

```python
sns.scatterplot(x="day", y="tip", hue="sex", data=tips)
```

```python
sns.barplot(x="day", y="tip", hue="sex", data=tips)
```
