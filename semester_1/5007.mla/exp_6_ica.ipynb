```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from sklearn.mixture import GaussianMixture
from scipy.io import wavfile

rate, mixed_audio = wavfile.read("./data/mixed_audio.wav")  
ica = FastICA(n_components=2)
sources = ica.fit_transform(mixed_audio)
gmm = GaussianMixture(n_components=2).fit(sources)
labels = gmm.predict(sources)

# Visualization
plt.scatter(sources[:, 0], sources[:, 1], c=labels, cmap='coolwarm')
plt.title("Separated Audio Sources using ICA and GMM")
plt.xlabel("Source 1")
plt.ylabel("Source 2")
plt.show()
```
